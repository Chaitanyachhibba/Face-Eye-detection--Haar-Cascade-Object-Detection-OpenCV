```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_eye.xml')
```

```python
cap = cv2.VideoCapture(0)
```

```python
img = cv2.imread('C:\\Users\\chait\\OneDrive\\Pictures\\Camera Roll\\1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

## Face Detection

```python
def detect_face(img):    
  
    face_img = img.copy()
  
    face_rects = face_cascade.detectMultiScale(face_img) 
    
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,0,0), 10) 
        
    return face_img
```

```python
result = detect_face(img)
plt.imshow(result)
```

## Eye Cascade File

```python
def detect_eyes(img):
    
    face_img = img.copy()
  
    eyes = eye_cascade.detectMultiScale(face_img) 
        
    for (x,y,w,h) in eyes: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (0,255,0), 10) 
        
    return face_img
    
```

```python
result = detect_eyes(img)
plt.imshow(result,cmap='gray')
```

```python
def detect_eyes_face(img):
    
    face_img = img.copy()
  
    face = face_cascade.detectMultiScale(face_img) 
        
    for (x,y,w,h) in face: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (0,255,0), 10) 
        
        roi_face_img = face_img[y:y+h, x:x+w]
        
        eyes = eye_cascade.detectMultiScale(roi_face_img)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_face_img,(ex,ey),(ex+ew,ey+eh),(255,0,0),5)
        
    return face_img
```

```python
result = detect_eyes_face(img)
plt.imshow(result,cmap='gray')
```
